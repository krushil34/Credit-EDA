### UNIVARIATE ANALYSIS
- analysing the client who paid and who didn't pay loan

```python
def graph_plot(var):
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    ax1=good_client[var].value_counts().plot(kind='bar',ax=ax1)
    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)

    abc=round(((good_client[var].value_counts() / 282076) * 100),2)
    for i,v in enumerate(abc):
        ax1.text(i, v-40, str(v)+'%', horizontalalignment='center', size=10, color='black',fontweight='bold')
    



    ax2=bad_client[var].value_counts().plot(kind='bar',ax=ax2)
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)

    bcd=round(((bad_client[var].value_counts() / 24773) * 100),2)
    for i,v in enumerate(bcd):
        ax2.text(i, v-4,str(v)+'%', horizontalalignment='center', size=10, color='black',fontweight='bold')
    
    plt.show()
```

```python
def graph_plot_bivariate(var1,var2):
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    sns.scatterplot(data=good_client,x=var1,y=var2,ax=ax1)
    ax1.set_title(f'Distribution of {var1} vs {var2} for Non-Defaulters',fontsize=15)

    # abc=round(((good_client[var].value_counts() / 282076) * 100),2)
    # for i,v in enumerate(abc):
    #     ax1.text(i, v-40, str(v)+'%', horizontalalignment='center', size=10, color='black',fontweight='bold')

    sns.scatterplot(data=bad_client,x=var1,y=var2,ax=ax2)
    ax2.set_title(f'Distribution of {var1} vs {var2} for Defaulters',fontsize=15)

    # bcd=round(((bad_client[var].value_counts() / 24773) * 100),2)
    # for i,v in enumerate(bcd):
    #     ax2.text(i, v-4,str(v)+'%', horizontalalignment='center', size=10, color='black',fontweight='bold')
    
    plt.show()
```
